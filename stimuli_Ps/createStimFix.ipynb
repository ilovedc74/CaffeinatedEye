```python
param_dict = { 
    "B": [r"/", "\\"], 
    "SOA":["10", "40", "110", "220", "500"], 
    "Ecc":["+6", "-6"], 
    "Con":["250"],
  }
```

```python
combinations = it.product(*(param_dict[Name] for Name in param_dict))
combinations = list(combinations)
for idx in range(len(combinations)):
    tmp = combinations[idx]
    combinations[idx] = '.'.join( 
        ["B="+tmp[0], "SOA="+tmp[1], "Ecc="+tmp[2], "Con="+tmp[3]]
    ) + r'.'
paramContainer = combinations * 5 # a list of len=100 with 5*20 all combinations
random.shuffle(paramContainer) #randomised final param list
paramContainer
```

```python
# combine it with the randomized content variable (take 200 from len(content))
# 1. classify rows into two categories: Eonlyrow or EQcombinerow
contentEonly = []
contentEQ = []
sep = '~'
for idx in range(len(content)):
    tmp = re.match('Q\.\d+.*', content[idx])
    if tmp == None:
        if idx == len(content)-1: # last case
            contentEonly.append(  content[idx])
        elif re.match('Q\.\d+.*', content[idx+1]) == None: # next also not Q sentence.
            contentEonly.append(  content[idx]  )
        else:
            contentEQ.append( content[idx+1] + sep + content[idx] ) # QE
    else:
        continue

```

```python
eitemNum = re.compile('E\.\d+\.')
tmp = re.compile('B=')


for idx in range(len(finalPre)):
    temp = eitemNum.search(finalPre[idx])[0]
    
    Bslash = tmp.search(finalPre[idx])[0]
    
    finalPre[idx] = re.sub(temp,'', finalPre[idx])            #delete
    finalPre[idx] = re.sub(Bslash,temp+Bslash, finalPre[idx]) #replace
finalPre[idx]
    
```
